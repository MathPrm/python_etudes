# La Fonction range() en Python

La fonction `range()` est un outil fondamental en Python qui permet de générer une séquence de nombres. Elle est particulièrement utile dans les boucles et pour créer des listes de nombres.

## 1. Range Simple - Un seul paramètre

La forme la plus simple de `range()` prend un seul argument qui représente la valeur de fin (non incluse). La séquence commence toujours par 0.

```python
# Exemple avec range(5)
print("Liste des nombres de 0 à 4 :")
for nombre in range(5):
    print(nombre, end=" ")

# Convertir en liste pour voir tous les éléments d'un coup
print("\n\nMême chose en liste :")
print(list(range(5)))
```

## 2. Range avec Début et Fin

On peut spécifier une valeur de début et de fin avec `range(début, fin)`. La séquence inclura le nombre de début mais pas le nombre de fin.

```python
# Exemple avec range(2, 7)
print("Liste des nombres de 2 à 6 :")
for nombre in range(2, 7):
    print(nombre, end=" ")

print("\n\nEn liste :")
print(list(range(2, 7)))
```

## 3. Range avec un Pas Personnalisé

Le troisième paramètre de `range(début, fin, pas)` permet de définir l'intervalle entre chaque nombre. Par défaut, le pas est de 1, mais on peut le modifier selon nos besoins.

```python
# Exemple avec un pas de 2
print("Nombres pairs de 0 à 10 :")
for nombre in range(0, 11, 2):
    print(nombre, end=" ")

print("\n\n# Exemple avec un pas négatif")
print("Compte à rebours de 10 à 0 :")
for nombre in range(10, -1, -1):
    print(nombre, end=" ")
```

## 4. Utilisation Pratique de Range

Voici quelques exemples pratiques d'utilisation de `range()` dans différentes situations :

```python
# Création d'une liste de carrés
carres = [n**2 for n in range(5)]
print("Liste des carrés de 0 à 4:", carres)

# Création d'une table de multiplication
nombre = 7
print(f"\nTable de multiplication de {nombre} :")
for i in range(1, 11):
    print(f"{nombre} × {i} = {nombre * i}")

# Exemple avec un index dans une boucle
fruits = ["pomme", "banane", "orange", "kiwi"]
print("\nListe de fruits avec leur index :")
for i in range(len(fruits)):
    print(f"Index {i}: {fruits[i]}")
```
